```python
def add_question(question, answer, explanation):
    global questions_df
    new_id = len(questions_df) + 1
    new_row = pd.DataFrame([{
        "QuestionID": new_id, "Question": question, "Answer": answer,
        "Explanation": "\n"+explanation, "Counter": 0  # Initialize counter to 0
    }])
    questions_df = pd.concat([questions_df, new_row], ignore_index=True)

def update_question(question_id, question=None, answer=None, explanation=None):
    global questions_df
    if question_id in questions_df["QuestionID"].values:
        if question is not None:
            questions_df.loc[questions_df["QuestionID"] == question_id, "Question"] = question
        if answer is not None:
            questions_df.loc[questions_df["QuestionID"] == question_id, "Answer"] = answer
        if explanation is not None:
            questions_df.loc[questions_df["QuestionID"] == question_id, "Explanation"] = explanation
    else:
        print("Question ID not found.")

def ask_question():
    global questions_df
    available_questions = questions_df[questions_df["Counter"] == 0]

    # Decrease counter for all questions that have been answered correctly previously
    questions_df.loc[questions_df["Counter"] > 0, "Counter"] -= 1
    
    if available_questions.empty:
        questions_df.loc[questions_df["Counter"] > 0, "Counter"] = 0
        print("No more questions available.")
        return
    
    question_row = available_questions.sample(n=1).iloc[0]
    question_id = question_row["QuestionID"]
    question = question_row["Question"].strip()
    explanation = question_row["Explanation"].strip()
    
    print(f'#{question_id} {question}')
    your_answer = input("Your answer: ")
    correct = your_answer.strip().lower() == question_row["Answer"].strip().lower()
    
    if correct:
        # Set counter to 50 if answer is correct
        questions_df.loc[questions_df["QuestionID"] == question_row["QuestionID"], "Counter"] = 50
        print("\nCorrect!")
    else:
        print("\nWrong. The correct answer is:", question_row["Answer"].strip())
    print(f"\nExplanation:\n{explanation}" )
```

```python
ask_question()
```
